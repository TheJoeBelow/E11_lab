```python
import numpy as np
import math
import pandas as pd
# This line allows plots to show in the Jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt
import io
import requests
```

```python
data1 = pd.read_csv('ourdata1.csv')
data2 = pd.read_csv('ourdata2.csv')
hdata1 = pd.read_csv('hanksdata1_2021-09-04_D3S.csv')
hdata2 = pd.read_csv('hanksdata2_2021-09-04_D3S.csv')
```

```python
temp_data = data1.loc[:,"CP5S":"Time"]
plt.hist(temp_data['CP5S'] ,bins = 7)
plt.ylabel('Frequency') # label the y-axis
plt.xlabel('Counts per 5 seconds')
plt.title("Statistical data for our detector")                                  # put a title!
plt.show()
ourmean1 = np.mean(temp_data['CP5S'])
ourstd1 = np.std(temp_data['CP5S'])
print(ourmean1,ourstd1)
```

```python
temp_data = data2.loc[:,"CP5S":"Time"]
plt.hist(temp_data['CP5S'] ,bins = 7)
plt.ylabel('Frequency') # label the y-axis
plt.xlabel('Counts per 5 seconds')
plt.title("Statistical data for our detector")                                  # put a title!
plt.show()
ourmean2 = np.mean(temp_data['CP5S'])
ourstd2 = np.std(temp_data['CP5S'])
print(ourmean2,ourstd2)
```

```python
#temp_data = hdata1.loc[:,"CPM":"Time"]
hdata1.to_numpy()
hcounts1 = np.sum(hdata1, axis=1)
plt.hist(hcounts1 ,bins = 50)
plt.ylabel('Frequency') # label the y-axis
plt.xlabel('Counts per 5 seconds')
plt.title("Statistical data for our detector")                                  # put a title!
plt.show()
mean1 = np.mean(hcounts1)
std1 = np.std(hcounts1)
print(mean1,std1)
```

```python
hdata2.to_numpy()
hcounts2 = np.sum(hdata2, axis=1)
plt.hist(hcounts2 ,bins = 15)
plt.ylabel('Frequency') # label the y-axis
plt.xlabel('Counts per 5 seconds')
plt.title("Statistical data for our detector")                                  # put a title!
plt.show()
mean2 = np.mean(hcounts2)
std2 = np.std(hcounts2)
print(mean2,std2)
```

### For the 2nd data with both our detector and the given detector, the standard deviation was approximately the square root of the mean, which fits with our expected values. However, with the unknown sample, the standard deviation was much more than the square root of the mean value.

```python
ourpercentunknown = ourstd1/ourmean1*100
ourpercentthor = ourstd2/ourmean2*100
hankspercentunknown = std1/mean1*100
hankspercentthor = std2/mean2*100

print("The percent errors with our detector for unknown souce and thorium respectively were {:.1f}%, {:.1f}%".format(ourpercentunknown, ourpercentthor))
print("The percent errors with the class detector for unknown souce and thorium respectively were {:.1f}%, {:.1f}%".format(hankspercentunknown, hankspercentthor))
```

```python
activity = 30593
oureff = ourmean2/activity
hankseff = mean2/activity
print("The efficiency of our detector and the class detector is {:.2e}, {:.2e} respectively".format(oureff,hankseff))
```

### 5) The class detector is more sensitive, it has twice the efficiency and is more sensitive to changes due to the higher overall counts.

```python
ouract = ourmean1/oureff
hanksact = mean1/hankseff
print("With our detector the activity of the sample is {:.0f} Bq".format(ouract))
print("With the class detector the activity of the sample is {:.0f} Bq".format(hanksact))
```
