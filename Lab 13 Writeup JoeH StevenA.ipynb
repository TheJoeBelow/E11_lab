```python
import numpy as np
import math
import pandas as pd
# This line allows plots to show in the Jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt
import io
import requests
```

```python
joedata = pd.read_csv('joenov19-2021.csv')
stevedata = pd.read_csv('steven-nov-19-2021.csv')
```

```python
joes_temp_rad = joedata.loc[:,'Radiation (cps)']
joes_temp_air = joedata.loc[:, 'Air Quality PM 2.5 (ug/m3)']
joetime = joedata.loc[:, 'Epoch time']
tot_time= [0]
for i in range(1,155):
    tot_time.append(round(joetime[i]- joetime[1]))
plt.plot(tot_time,joes_temp_rad)
plt.ylabel('Counts per second') # label the y-axis
plt.xlabel('Time (s)')
plt.title("Radiation data taken over an approximately 25 minute walk around UC Berkeley.(Joe)")                                  # put a title!
plt.show()
plt.plot(tot_time,joes_temp_air)
plt.ylabel('Air Quality (ug/m3)') # label the y-axis
plt.xlabel('Time (s)')
plt.title("Air Quality data taken over an approximately 25 minute walk around UC Berkeley. (Joe)")                                  # put a title!
plt.show()
plt.hist(joes_temp_rad ,bins = 20)
plt.ylabel('Frequency') # label the y-axis
plt.xlabel('Counts per second')
plt.title("Variation on Counts per Second (Joe)")                                  # put a title!
plt.show()
```

```python
steves_temp_rad = stevedata.loc[:,'Radiation (cps)']
steves_temp_air = stevedata.loc[:, 'Air Quality PM 2.5 (ug/m3)']
stevetime = stevedata.loc[:, 'Epoch time']
tot_time= [0]
for i in range(1,154):
    tot_time.append(round(stevetime[i]- stevetime[1]))
plt.plot(tot_time,steves_temp_rad)
plt.ylabel('Counts per second') # label the y-axis
plt.xlabel('Time (s)')
plt.title("Radiation data taken over an approximately 25 minute walk around UC Berkeley. (Steve)")                                  # put a title!
plt.show()
plt.plot(tot_time,steves_temp_air)
plt.ylabel('Air Quality (ug/m3)') # label the y-axis
plt.xlabel('Time (s)')
plt.title("Air Quality data taken over an approximately 25 minute walk around UC Berkeley. (Steve)")                                  # put a title!
plt.show()
plt.hist(steves_temp_rad ,bins = 20)
plt.ylabel('Frequency') # label the y-axis
plt.xlabel('Counts per second')
plt.title("Variation on Counts per Second (Steve)")                                  # put a title!
plt.show()
```

```python
joemean = np.mean(joes_temp_rad)
stevemean = np.mean(steves_temp_rad)
joestd = np.std(joes_temp_rad)
stevestd = np.std(steves_temp_rad)
print("The mean values for counts per second were {:.2f} and {:.2f} for Joe and Steven respectively".format(joemean,stevemean))
print("The standard deviations for counts per second were {:.2f} and {:.2f} for Joe and Steven respectively".format(joestd,stevestd))
```

These values agree with each other statistically.

```python
print("The uncertainties for cps based soley on counting statistics should be {:.2f} and {:.2f} for Joe and Steven respectively".format(math.sqrt(joemean),math.sqrt(stevemean)))
```

Our standard deviations were slightly less than this but not vastly different.

```python
joesuncert = math.sqrt(joemean)/(math.sqrt(2*math.sqrt(155)))
stevesuncert = math.sqrt(stevemean)/(math.sqrt(2*math.sqrt(154)))
print("The uncertainty on the uncertainty of our data is {:.2f} and {:.2f} for Joe and Steven respectively".format(joesuncert,stevesuncert))
```

```python
print("With this new evaluation, Joes uncertainty becomes {:.2f} + or - {:.2f}.".format(joestd,joesuncert))
print("and Stevens uncertainty becomes {:.2f} + or - {:.2f}.".format(stevestd,stevesuncert))
```

With these taken into consideration, we have high confidence in our experiment. We also have confidence in the levels of radiation we were detecting, since they errors fall within the expected ranges for counting statistics.
