```python
import numpy as np
import math
import pandas as pd
# This line allows plots to show in the Jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt
import io
import requests
```

```python
idata = pd.read_csv('inside_data.csv')
odata = pd.read_csv('outside_data.csv')
```

```python
#print(data)
```

```python
temp_data = idata.loc[:,"Temperature":"Time"]
print(temp_data)
```

```python
plt.plot(temp_data['Temperature'])
plt.ylabel('Temperature (C)') # label the y-axis
plt.xlabel('Time (s)')
plt.title("Temperature against the progression of time")                                  # put a title!
plt.show()
plt.hist(temp_data['Temperature'] ,bins = 10)
plt.ylabel('Frequency') # label the y-axis
plt.xlabel('Temperatures (C)')
plt.title("Range of Temperatures")                                  # put a title!
plt.show()
```

```python
plt.plot(temp_data['Humidity'])
plt.ylabel('Humidity (gH20/kg air)') # label the y-axis
plt.xlabel('Time (s)')
plt.title("Changes in Humidity in Jacobs Basement")                                  # put a title!
plt.show()
plt.hist(temp_data['Humidity'] ,bins = 20)
plt.ylabel('Frequency') # label the y-axis
plt.xlabel('Humidity (gH20/kg air)')
plt.title("Frequency of Humidities in Jacobs basement")                                  # put a title!
plt.show()
```

```python
plt.plot(temp_data['Pressure'])
plt.ylabel('Pressure (hPa)')
plt.xlabel('Time (s)')
plt.title("Under Pressure")                                  
plt.show()
plt.hist(temp_data['Pressure'] ,bins = 10)
plt.ylabel('Frequency') # label the y-axis
plt.xlabel('Pressure (hPa)')
plt.title("Frequency of Pressuress in Jacobs basement")                                  # put a title!
plt.show()
```

```python

plt.plot(temp_data['PM1'], label = 'PM1')
plt.plot(temp_data['PM2.5'], label = 'PM2.5')
plt.plot(temp_data['PM10'], label = 'PM10')
plt.legend()
plt.ylabel('Particles')
plt.xlabel('Time (s)')
plt.title("CPM vs index")                                  
plt.show()
```
